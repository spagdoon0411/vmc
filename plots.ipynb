```python
import pickle
```

```python
psi_100 = pickle.load(open("psi_100_reads_10000_iters.pkl", "rb"))
```

```python
psi_1000 = pickle.load(open("psi_1000_reads_1000_iters.pkl", "rb"))
```

```python
psi_100
```

```python
psi_100_avg = psi_100.mean(axis=1)
```

```python
psi_1000_avg = psi_1000.mean(axis=1)
```

```python
import numpy as np
```

```python
psi_100_avg_norm = psi_100_avg / np.sqrt((psi_100_avg ** 2).sum())
```

```python
psi_1000_avg_norm = psi_1000_avg / np.sqrt((psi_1000_avg ** 2).sum())
```

```python
import matplotlib.pyplot as plt
plt.plot(psi_100_avg_norm, label="100 reads")
```

```python
plt.plot(psi_1000_avg_norm, label="1000 reads")
```
